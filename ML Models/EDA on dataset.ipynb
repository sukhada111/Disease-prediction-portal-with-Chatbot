```python
#checking the data distribution for some features
features=['pain_behind_the_eyes','redness_of_eyes','sinus_pressure','runny_nose','congestion','chest_pain',
            'weakness_in_limbs','fast_heart_rate','pain_during_bowel_movements']

plt.figure(figsize=(17,17))

for i,feature in enumerate(features):
    plt.subplot(4,3,i+1)
    sns.countplot(x=feature,data=df)
    plt.xlabel(feature)
    
plt.show()
```

```python
features=['indigestion','headache','yellowish_skin','dark_urine','nausea','loss_of_appetite','back_pain','constipation',
          'abdominal_pain','diarrhoea','mild_fever']

plt.figure(figsize=(17,17))

for i,feature in enumerate(features):
    plt.subplot(4,3,i+1)
    sns.countplot(x=feature,data=df)
    plt.xlabel(feature)
    
plt.show()
```

```python
#plotting heatmap to find the correlation between features

df_corr=df.iloc[:,10:40]
plt.figure(figsize=(30, 30))

sns.heatmap(df_corr.corr(),annot = True)

plt.show()
```
